# Problem Statement:
Consider yourself to be Sam who is a data scientist. He has been approached by

# Tasks To Be Performed:
1. Sam has to build a bar-plot for the ‘Contract’ column:
<br>a. Set the x-axis label to be ‘Contract Type of customer’
<br>b. Set the y-axis label to be ‘Count’
<br>c. Set the title of the plot to be ‘Distribution of Contract’
<br>d. Assign ‘orange’ color to all the bars<br><br>
2. Sam has to build a histogram for the ‘MonthlyCharges’ column:
<br>a. Set the x-axis label to be ‘Monthly Charges Incurred’
<br>b. Set the y-axis label to be ‘Count’
<br>c. Set the title of the plot to be ‘Distribution of Monthly Charges’
<br>d. Assign ‘forestgreen’ color to the bins<br><br>
3. Sam has to build a scatter-plot between ‘TotalCharges’ and ‘tenure’.
‘TotalCharges’ should be on the y-axis and ‘tenure’ should be on the
x-axis.
<br>a. Set the x-axis label to be ‘Tenure of the customer’
<br>b. Set the y-axis label to be ‘Total chargesIncurred’
<br>c. Set the title of the plot to be ‘Total Charges vs Tenure’
<br>d. Assign ‘indigo’ color to the points<br><br>
4. Sam has to build a box-plot between ‘MonthlyCharges’ and
‘PaymentMethod’. ‘MonthlyCharges’ should be on the y-axis and
‘PaymentMethod’ should be on the x-axis.
<br>a. Set the x-axis label to be ‘Payment Method of customer’
<br>b. Set the y-axis label to be ‘Monthly ChargesIncurred’
<br>c. Set the title of plot to be ‘Monthly Charges vs. Payment Method’
<br>d. Assign ‘olive’ color to the box-plots

```python
df = pd.read_csv("/content/customer_churn-1.csv",na_values=" ")
df.columns
```

```python
# 1
sns.countplot(x=df.Contract,color='orange')
plt.xlabel("Contract Type of customer")
plt.ylabel("Count")
plt.title("Distribution of Contract")
plt.show()
```

```python
# 2
sns.histplot(x=df.MonthlyCharges,color='forestgreen',kde=True)
plt.xlabel('Monthly Charges Incurred')
plt.ylabel('Count')
plt.title('Distribution of Monthly Charges')
plt.show()
```

```python
# 3
df.TotalCharges.astype(dtype=float)
plt.figure(figsize=(10,8))
sns.scatterplot(y=df.TotalCharges,x=df.tenure,color='indigo')
plt.xlabel('Tenure of the customer')
plt.ylabel('Total Charges Incured')
plt.title('Total Charges vs Tenure')
plt.show()
```

```python
# 4
plt.figure(figsize=(10,5))
sns.boxplot(x=df.PaymentMethod,y=df.MonthlyCharges,color='olive')
plt.xlabel('Payment Method of customer')
plt.ylabel('Monthly Charges Incurred')
plt.title('Monthly Charges vs Payment Method')
plt.show()
```
